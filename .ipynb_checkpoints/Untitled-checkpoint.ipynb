```python
import imdlib as imd
import matplotlib.pyplot as plt

# File configuration
file_path = "Rainfall_ind2025_rfp25.grd"

# Read the specialized IMD binary format (0.25-degree resolution)
# Note: file_dir must point to the folder containing your file
data = imd.open_data(var_type='rain', start_yr=2025, end_yr=2025, fn_format='single', file_dir='./')

# Convert it into an Xarray object
ds = data.get_xarray()

# Print the data structure (Dimensions: lat: 129, lon: 135, time: 365)
print(ds)

# Quick validation plot of Day 1 rainfall over India
ds.to_array().isel(time=0).plot(cmap="Blues", robust=True)
plt.title("IMD Daily Rainfall - Jan 1, 2025")
plt.show()
```

```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt

# 1. Convert imdlib object to Xarray Dataset (Assuming variable name is 'rain')
# If your variable name differs, check your printed ds structure from the previous step
grid_ds = ds

# 2. Spatial Aggregation: Calculate Country-Wide Daily Average Time Series
country_daily_avg = grid_ds['rain'].mean(dim=['lat', 'lon'])

# 3. Temporal Aggregation: Calculate Total Accumulated Rainfall Per Grid Point
total_rainfall = grid_ds['rain'].sum(dim='time')

# 4. Find the Peak Rainfall Event Day
peak_day_index = int(country_daily_avg.argmax())
peak_date = pd.to_datetime(country_daily_avg.time[peak_day_index].values)
peak_value = float(country_daily_avg[peak_day_index].values)

print(f"--- 2025 DATASET OVERVIEW ---")
print(f"Max daily average rainfall across India: {peak_value:.2f} mm occurred on {peak_date.strftime('%B %d, %Y')}")
print(f"Total annual average grid rainfall: {float(total_rainfall.mean()):.2f} mm")
```

```python
plt.figure(figsize=(12, 5))
country_daily_avg.plot(color="navy", linewidth=1.5)
plt.title("India Country-Wide Daily Average Rainfall Timeline (2025)", fontsize=14, pad=15)
plt.xlabel("Date")
plt.ylabel("Average Rainfall (mm/day)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Assuming 'ds' is your Xarray Dataset from imdlib
# Convert Xarray to a flattened Pandas Dataframe
df = ds.to_dataframe().reset_index()

# Clean missing values (IMD uses 99.9 or -99.9 for non-land/ocean areas)
df = df[df['rain'] >= 0] 

# Feature Engineering: Extract time elements for the ML model
df['time'] = pd.to_datetime(df['time'])
df['day_of_year'] = df['time'].dt.dayofyear
df['month'] = df['time'].dt.month

# Final features: Predict 'rain' using spatial coordinates and time variables
X = df[['lat', 'lon', 'day_of_year', 'month']]
y = df['rain']
```

```python
# Use the first 80% of days for training, last 20% for testing
unique_dates = df['time'].unique()
split_idx = int(len(unique_dates) * 0.8)
train_dates = unique_dates[:split_idx]

train_mask = df['time'].isin(train_dates)

X_train, y_train = X[train_mask], y[train_mask]
X_test, y_test = X[~train_mask], y[~train_mask]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the model
rf_model = RandomForestRegressor(n_estimators=100, max_depth=12, random_state=42, n_jobs=-1)

# Train the model
rf_model.fit(X_train, y_train)
```

```python
# Make predictions on the test partition
y_pred = rf_model.predict(X_test)

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"✅ Training Complete")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mse):.3f} mm")
print(f"R² Score: {r2:.3f}")
```

```python
import numpy as np
import pandas as pd
import xarray as xr

# 1. IMD 0.25-degree grid specs (change size if you use 1-degree or 0.5-degree)
n_lat, n_lon = 129, 135  
latitudes = np.linspace(6.5, 38.5, n_lat)
longitudes = np.linspace(66.5, 100.0, n_lon)

all_datasets = []

# 2. Loop and construct the grids manually
for path in file_paths:
    # Read flat binary data (IMD uses 4-byte floats)
    raw_data = np.fromfile(path, dtype=np.float32)
    
    # Calculate how many days are inside this file (e.g., 365 or 366)
    days = len(raw_data) // (n_lat * n_lon)
    
    # Reshape to (time, lat, lon)
    reshaped = raw_data.reshape((days, n_lat, n_lon))
    
    # Handle missing values (IMD uses -999.0 for missing data)
    reshaped[reshaped == -999.0] = np.nan
    
    # Create fake or parsed time array for this specific file
    # (Adjust '2024-01-01' to match the actual year of the file)
    time_coords = pd.date_range(start="2024-01-01", periods=days, freq="D")
    
    # Build the xarray object
    da = xr.DataArray(
        reshaped,
        coords=[time_coords, latitudes, longitudes],
        dims=["time", "latitude", "longitude"],
        name="rainfall",
    )
    all_datasets.append(da.to_dataset())

# 3. Combine your multi-year datasets together cleanly!
ds = xr.concat(all_datasets, dim="time")
print(ds)
```

```python
import os
import glob
import numpy as np
import pandas as pd
import xarray as xr
import rioxarray
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define the path to your folder containing .grd files
DATA_FOLDER = r".\rainfall_data"
```

```python
import os
import glob
import numpy as np
import pandas as pd

# Define grid size for IMD 0.25 degree data
COLS = 135  # Longitude (66.5E to 100.0E)
ROWS = 129  # Latitude (6.5N to 38.5N)
GRID_SIZE = COLS * ROWS

# Find all files
grd_files = sorted(glob.glob(os.path.join(DATA_FOLDER, "*.grd")))

data_list = []
all_timestamps = []

print(f"Found {len(grd_files)} IMD binary files. Processing...")

for file_path in grd_files:
    filename = os.path.basename(file_path)
    
    # Extract the year from the filename (e.g., 'Rainfall_ind1995_rfp25.grd' -> 1995)
    # Adjust slicing if your naming structure varies slightly
    try:
        year = int("".join(filter(str.isdigit, filename))[:4])
    except ValueError:
        print(f"Skipping {filename}: Could not determine year.")
        continue
        
    # Determine the number of days in that year
    is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
    num_days = 366 if is_leap else 365
    
    # Read the raw binary data
    try:
        with open(file_path, 'rb') as f:
            # IMD data uses 4-byte float32
            raw_data = np.fromfile(f, dtype=np.float32)
            
        # Verify the file size matches expected days * grid dimensions
        expected_size = num_days * GRID_SIZE
        if raw_data.size != expected_size:
            print(f"Warning: {filename} has unexpected size. Found {raw_data.size}, expected {expected_size}. Sub-setting or skipping...")
            # If the file contains fewer/more records, adjust num_days dynamically
            num_days = raw_data.size // GRID_SIZE
            raw_data = raw_data[:num_days * GRID_SIZE]
            
        # Reshape data into (Days, Rows, Columns)
        yearly_grid = raw_data.reshape((num_days, ROWS, COLS))
        
        # Handle IMD missing data code (usually -99.9 or similar negative values)
        # You can set it to 0 or NaN depending on your preference
        yearly_grid[yearly_grid < 0] = np.nan
        
        data_list.append(yearly_grid)
        
        # Create explicit time indices for this specific year
        year_dates = pd.date_range(start=f"{year}-01-01", periods=num_days, freq='D')
        all_timestamps.extend(year_dates)
        
        print(f"Successfully loaded {year} ({num_days} days).")
        
    except Exception as e:
        print(f"Failed to parse {filename}: {e}")

# Stack all years together into one continuous timeline
stacked_data = np.concatenate(data_list, axis=0)
n_times, height, width = stacked_data.shape

print("---")
print(f"Final Data Shape: {stacked_data.shape} (Total Days: {n_times}, Lat Rows: {height}, Lon Cols: {width})")
```

```python
# 1. Create a mask of land pixels (where data is not always NaN)
# If a pixel is NaN for all time steps, it's likely ocean/outside borders
land_mask = ~np.isnan(stacked_data).all(axis=0) 
n_land_pixels = np.sum(land_mask)

print(f"Total grid pixels: {height * width}")
print(f"Active land pixels to train on: {n_land_pixels}")

# 2. Extract only the land pixels over time
# Shape changes from (Days, Lat, Lon) to (Days, Land_Pixels)
flat_land_data = stacked_data[:, land_mask]

# 3. Fill sporadic missing grid values on land using interpolation
flat_land_data = pd.DataFrame(flat_land_data).interpolate(method='linear', limit_direction='both').values

# 4. Create sliding window features (e.g., lookback of 7 days)
lookback = 7
X, y = [], []

for i in range(len(flat_land_data) - lookback):
    X.append(flat_land_data[i : i + lookback].flatten()) # Past 7 days rolled into 1 vector
    y.append(flat_land_data[i + lookback])              # The 8th day

X = np.array(X)
y = np.array(y)

print(f"Features shape (X): {X.shape} (Samples, Lookback * Land_Pixels)")
print(f"Target shape (y): {y.shape} (Samples, Land_Pixels)")
```

```python
from xgboost import XGBRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error, r2_score
import time

# 1. Chronological Split (80% Train, 20% Test)
split_idx = int(len(X) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

print(f"Training Data Shape: {X_train.shape}")
print("Initializing XGBoost Multi-Output Regressor...")

# 2. Define base XGBoost parameters optimized for time-series regression
base_xgb = XGBRegressor(
    n_estimators=100,      # Number of boosting rounds per pixel
    max_depth=6,           # Depth of trees to capture regional interactions
    learning_rate=0.05,    # Step size shrinkage to prevent overfitting
    subsample=0.8,         # Row subsampling for variance reduction
    colsample_bytree=0.8,  # Column subsampling
    tree_method='hist',    # Fast histogram optimized framework
    random_state=42,
    n_jobs=-1              # Use all available CPU cores
)

# 3. Wrap it to handle multiple pixels
xgb_multi_model = MultiOutputRegressor(base_xgb)

# 4. Train the model (This might take a few minutes depending on CPU)
start_time = time.time()
print("Training started...")
xgb_multi_model.fit(X_train, y_train)
print(f"Training completed in {time.time() - start_time:.2f} seconds.")

# 5. Evaluate performance
y_pred = xgb_multi_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("--- Evaluation Results ---")
print(f"Test Root Mean Squared Error (RMSE): {rmse:.4f} mm")
print(f"Test R² Score: {r2:.4f}")
```
